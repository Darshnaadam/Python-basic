```python
# Adding parameters to the class
```

```python
class phone:
    def set_color(self,color):
        self.color=color
        
    def set_cost(self,cost):
        self.cost=cost
        
    def show_color(self):
        return self.color
    
    def show_cost(self):
        return self.cost
    
    def play_game(self):
        print("Playing a game")
        
    def make_call(self):
        print("Make a call")
```

```python
p2 = phone()
```

```python
p2.set_color('blue')
p2.set_cost(500)
```

```python
p2.show_color()
```

```python
p2.show_cost()
```

```python
p2.play_game()
```

```python
p2.make_call()
```

```python
# Creating a class with constructor
```

```python
class Employee:
    
    def __init__(self,name,age,salary,gender):
        self.name = name
        self.age = age
        self.salary = salary
        self.gender = gender
        
    def employee_details(self):
        print("Name of the Employee is ",self.name)
        print("Age of the Employee is ",self.age)
        print("Salary of the Employee is ",self.salary)
        print("Gender of the Employee is ",self.gender)
```

```python
e1 = Employee("Sam",18,25000,"male")
```

```python
e1.employee_details()
```
